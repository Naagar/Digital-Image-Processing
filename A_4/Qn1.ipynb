```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 
from scipy import misc  #here is how you get the racoon image

image = cv2.imread("Qn1.jpg",0)


def binary_racoon(image, lowerthreshold):
    img = image.copy()
    shape = np.shape(img)
    
    for i in range(shape[0]):
        for j in range(shape[1]): 
            if img[i,j] < lowerthreshold :
                     img[i][j] = 255
            else:
                 img[i][j] = 0

    return img

bina = binary_racoon(image,  127)
sh_p = np.shape(bina)
n = sh_p[0]
m = sh_p[1]
#plt.imshow(convertedpicture, cmap=plt.cm.gist_gray)
plt.subplot(111),plt.imshow(bina, cmap = 'gray')
plt.title('I-1 '), plt.xticks([]), plt.yticks([])
plt.show()
```

#  I-1(se1)---->I-2

```python
for i in range(2,n-1):
    for j in range(2,m-1):
        if bina[i][j] == 255:
            
            one_1 = ((bina[i][j])/255 + (bina[i+1][j-1])/255 + (bina[i+1][j])/255 + (bina[i+1][j+1])/255)
            zero_0 = bina[i-1][j-1]/255 + bina[i-1][j]/255 + bina[i-1][j+1]/255
        
            if one_1 == 4 and zero_0 == 0 :
                bina[i][j] = 0
            else:
                bina[i][j] = 255
plt.subplot(111),plt.imshow(bina, cmap = 'gray')
plt.title('I-2 '), plt.xticks([]), plt.yticks([])
plt.show()                
```
