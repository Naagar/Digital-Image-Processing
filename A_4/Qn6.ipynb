```python
import numpy as np
import matplotlib.pylab as plt


im = plt.imread('color_bars.tif')



def plti(im, h=8, **kwargs):
    
    y = im.shape[0]
    x = im.shape[1]
    w = (y/x) * h
    plt.figure(figsize=(w,h))
    plt.imshow(im, interpolation="none", **kwargs)
    plt.axis('off')

plti(im)

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

for c, ax in zip(range(3), axs):
    tmp_im = np.zeros(im.shape, dtype="uint8")
    tmp_im[:,:,c] = im[:,:,c]
    ax.imshow(tmp_im)
    ax.set_axis_off()
```

## (8-bit representation according to Safe RGB color model.)

```python
import numpy as np
import cv2
import matplotlib.pylab as plt
 
img  = cv2.imread('color_bars.tif')
b,g,r = cv2.split(img)
siz = np.shape(b)

saferplane = ((r/255.0)*5)*51
safegplane = ((g/255.0)*5)*51
safebplane = ((b/255.0)*5)*51

safeImage = np.empty(img.shape)
safeImage[:,:,0] = saferplane
safeImage[:,:,1] = safegplane
safeImage[:,:,2] = safebplane


plt.subplot(111),plt.imshow(safeImage, cmap = 'gray')
plt.title('Safe  8-bit RGB '), plt.xticks([]), plt.yticks([])
plt.show()
```

# Part-3 
## in this part 
## fisrt result/output is histogram of RGB image
## next three hisogram of Red,Green and Blue respectively
## next three histogram are Equalize histogram for Red,Green and Blue respectively

```python

import numpy as np
import cv2
import matplotlib.pylab as plt
 
img  = cv2.imread('color_bars.tif')    

b, g, r = cv2.split(img)

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()

equ_r = cv2.equalizeHist(r)
equ_g = cv2.equalizeHist(g)
equ_b = cv2.equalizeHist(b)

plt.hist(r.ravel(),256,[0,256]); 
plt.show()
plt.hist(g.ravel(),256,[0,256]); 
plt.show()
plt.hist(b.ravel(),256,[0,256]); 
plt.show()
plt.hist(equ_r.ravel(),256,[0,256]); 
plt.show()
plt.hist(equ_g.ravel(),256,[0,256]); 
plt.show()
plt.hist(equ_b.ravel(),256,[0,256]); 
plt.show()

```
