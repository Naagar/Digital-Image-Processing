```python
import cv2
import matplotlib.pyplot as plt 
image = cv2.imread('Qn2.jpg')
shifted = cv2.pyrMeanShiftFiltering(image, 21, 51)
```

```python
gray = cv2.cvtColor(shifted, cv2.COLOR_BGR2GRAY)
th = cv2.threshold(gray, 0, 255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
plt.subplot(111),plt.imshow(th, cmap = 'gray')
plt.title('Thresh '), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
cnts = cv2.findContours(th.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[-2]
for (i, c) in enumerate(cnts):
    ((x, y), _) = cv2.minEnclosingCircle(c)
    cv2.putText(image, "#{}".format(i + 1), (int(x) - 10, int(y)),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)
    cv2.drawContours(image, [c], -1, (0, 255, 0), 2)

plt.subplot(111),plt.imshow(image, cmap = 'gray')
plt.title('Labeled '), plt.xticks([]), plt.yticks([])
plt.show()
```
