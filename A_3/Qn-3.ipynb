```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('64lena.bmp',0)
f = np.fft.fft2(img) #fast fourier transformation
fshift = np.fft.fftshift(f) # shifting to center
Final = 20*np.log(np.abs(fshift)) # log scaled 

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(Final, cmap = 'gray')
plt.title(' FFT 64x64 Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt



img1 = cv2.imread('64lena.bmp',0)


constant= cv2.copyMakeBorder(img1,0,64,0,64,cv2.BORDER_CONSTANT,value=0)
f = np.fft.fft2(constant) #fast fourier transformation
fshift = np.fft.fftshift(f) # shifting to center
Final = 20*np.log(np.abs(fshift)) # log scaled 

#plt.subplot(131),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(121),plt.imshow(constant,'gray'),plt.title('Padded')
plt.subplot(122),plt.imshow(Final,'gray'),plt.title('Final FFT 128x128 Spectrum')


plt.show()
```
