# Qn-1(a)  

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('brick.tiff',0)
dft = np.fft.fft2(img) #fast fourier transformation
fshift = np.fft.fftshift(dft) # shifting to center
final = 20*np.log(np.abs(fshift)) # log scaled 

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(final, cmap = 'gray')
plt.title('Fourier Spectrum '), plt.xticks([]), plt.yticks([])
plt.show()
```

# Taking the conjugate, we can get the orignal image 

```python

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Brick-4.jpg',0)
f = np.fft.fft2(img) #fast fourier transformation
 
fshift = np.fft.fftshift(f) # shifting to center
final = 20*np.log(np.abs(fshift)) # log scaled 
f_of_fft = np.fft.fft2(final)  # fft of fft
fshift_fft = np.fft.fftshift(f_of_fft) # shifting to center
final_fft = 20*np.log(np.abs(fshift_fft))

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(final_fft, cmap = 'gray')
plt.title('Output Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
