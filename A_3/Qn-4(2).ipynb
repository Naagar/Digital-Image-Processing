```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal

imgF = cv2.imread('Brick-4.jpg',0)
#print("Before padding")
h_size = np.size(imgF)
#print(h_size)

imgH = cv2.imread('Bricks.jpg',0)
f_size = np.size(imgH)
#print(f_size,)
f_z= cv2.copyMakeBorder(imgF,0,50,0,50,cv2.BORDER_CONSTANT,value=0) #padding the image with smaller size
#print("After padding")
f_size = np.size(imgH)
#print(f_size)
h_size = np.size(f_z)
#print(h_size)


dftf_z = np.fft.fft2(f_z) #fast fourier transformation of f
dfth_z = np.fft.fft2(imgH)  #fast fourier transformation of h

dft_final = dftf_z*dfth_z # multplying the fft of f and fft of h
#f_z= cv2.copyMakeBorder(dft_final,0,50,0,50,cv2.BORDER_CONSTANT,value=0) #padding the image with smaller size


fshift = np.fft.fftshift(dft_final) # shifting to center for f*h
final = 20*np.log(np.abs(fshift)) # log scaled  for f*h
print(np.size(final))

con = signal.fftconvolve(dftf_z,dfth_z,mode='full') # finding the cocolution for f and h
print(np.size(con))
#fshift_con = np.fft.fftshift(con) # shifting to center for convolution
final_con = 20*np.log(np.abs(con)) # log scaled for convolution

plt.subplot(131),plt.imshow(f_z, cmap = 'gray')
plt.title('Image-Padded'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(final, cmap = 'gray')
plt.title('Fourier Spectrum '), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(final_con, cmap = 'gray')
plt.title('convolution '), plt.xticks([]), plt.yticks([])
plt.show()
```
