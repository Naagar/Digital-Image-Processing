```python
import cv2
import math
import copy
import numpy as np 
from matplotlib import pyplot as plt 
BLUE = [0, 0, 0]
img1 = cv2.imread('sky.jpeg', 0)
im= cv2.copyMakeBorder(img1,1,1,1,1,cv2.BORDER_CONSTANT,value=BLUE)
out_im = copy.copy(im)
for i in range(0, img1.shape[0]):
    for j in range(0, img1.shape[1]):
        ij = im[i-1,j-1] + im[i-1,j]*2 + im[i-1,j+1] + im[i,j-1]*2 + im[i,j]*4 + im[i,j+1]*2 + im[i+1,j-1] + im[i+1,j]*2 + im[i+1,j+1]
        out_im[i,j] = ij/16 
plt.subplot(121), plt.imshow(im, cmap='gray'), plt.title('Inital')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(out_im, cmap='gray'), plt.title('After Linear filter')
plt.xticks([]), plt.yticks([])
plt.show()
```
