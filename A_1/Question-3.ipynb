# Qn-3  piecewise transformation part(a)

```python
import cv2
import math
import copy
import numpy as np 
from matplotlib import pyplot as plt 
# read image 
im = cv2.imread('high3.jpg', 0)
# copy image to implement transformation
out_im = copy.copy(img)
# loop to implement peice wise transformation
for i in range(0, im.shape[0]):
    for j in range(0, im.shape[1]):
        if im[i,j] <= 110:
            out_im[i,j] = 2 * im[i,j] + 100
        elif im[i,j] <= 200:
            out_im[i,j] = im[i,j]
        else:
            out_im[i,j] = 255
plt.subplot(121), plt.imshow(im, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(out_im, cmap='gray'), plt.title('linear transformation')
plt.xticks([]), plt.yticks([])
plt.show()
```
