```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 

def BitQuantizeImage(im,k):
    k = 8-k
    im_new = np.zeros([im.shape[0],im.shape[1]],dtype = np.uint8)
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            im_new[i][j] = im[i][j]/(2**k)
    fig, (ax1,ax2) = plt.subplots(1,2) 
    ax1.imshow(im,"gray")
    ax2.imshow(im_new,"gray")
    plt.show()
    return im_new

img = cv2.imread("lena.bmp",0)

BitQuantizeImage(img,2)
BitQuantizeImage(img,3)
BitQuantizeImage(img,4)
BitQuantizeImage(img,5)
BitQuantizeImage(img,7)
```
